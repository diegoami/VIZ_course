```python
import pandas as pd
```

```python
df = pd.read_csv('../data/measles_incidence_per_year.csv', index_col=0)
```

```python
df['MEAN_V']=df.sum(axis=1)
```

```python
df
```

```python

import plotly as py
# and enable the offline mode in the notebook
py.offline.init_notebook_mode(connected=True)
# Building blocks of Plotly plots
import plotly.graph_objs as go
```

```python
trace0 = go.Scatter(x=df.index, y=df['MEAN_V'], name='Average Measles incidence per year', mode='lines+markers')


# Layout
layout = go.Layout(title='Average Measles incidence per year',
                   xaxis=dict(title='Year'),
                   yaxis=dict(title='average measles incidence'),
                  annotations=[{"x": 1963, "y" :50, "showarrow" : True, "arrowhead" : 0, "text" :"Vaccine <BR> introduced" , "ax" : 0, "ay" : -200} ])

# Figure
fig = go.Figure(data=[trace0], layout=layout) 
```

```python
trace = go.Heatmap(x=df.index.values, y=df.columns, z=df.values.T.tolist())

layout = go.Layout( 
   
    showlegend = False,
    height = 1000,width=1000,
       
    xaxis=dict(
         tickfont=dict(
            size=8,
        )
    ),
    yaxis=dict(
        tickfont=dict(
            size=8,
        ),
        
    ))
figure = dict(data=[trace],layout=layout)
```
